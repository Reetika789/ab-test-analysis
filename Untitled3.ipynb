```python
import pandas as pd

# Step 1: Load the CSV
csv_path = "ab_data.csv" # Corrected path to the uploaded CSV file
ab_data = pd.read_csv(csv_path)

# Step 2: Preview the data
print(ab_data.head())
```

```python
print(ab_data.columns)
print(ab_data.shape)
print(ab_data.head())
```

```python
# Total users in each group
n_old = ab_data[ab_data['group'] == 'control'].shape[0]
n_new = ab_data[ab_data['group'] == 'treatment'].shape[0]

# Total conversions in each group
conv_old = ab_data[(ab_data['group'] == 'control') & (ab_data['converted'] == 1)].shape[0]
conv_new = ab_data[(ab_data['group'] == 'treatment') & (ab_data['converted'] == 1)].shape[0]

print(f"Control group: {n_old} users, {conv_old} converted")
print(f"Treatment group: {n_new} users, {conv_new} converted")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Z-test to compare conversion rates
z_score, p_value = proportions_ztest([conv_new, conv_old], [n_new, n_old], alternative='larger')

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.6f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

# Conversion rates
conv_rate_old = conv_old / n_old
conv_rate_new = conv_new / n_new

# Confidence intervals
ci_old = sm.stats.proportion_confint(conv_old, n_old, alpha=0.05, method='normal')
ci_new = sm.stats.proportion_confint(conv_new, n_new, alpha=0.05, method='normal')

# Plotting
groups = ['Control (Old)', 'Treatment (New)']
rates = [conv_rate_old, conv_rate_new]
errors = [conv_rate_old - ci_old[0], conv_rate_new - ci_new[0]]

plt.figure(figsize=(8, 5))
plt.bar(groups, rates, yerr=errors, capsize=10, color=['skyblue', 'salmon'])
plt.ylabel('Conversion Rate')
plt.title('Conversion Rates with 95% Confidence Intervals')
plt.ylim(0, max(rates) + 0.02)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Z-score: -1.24
# P-value: 0.891942
# Fail to reject H₀: No significant improvement from new design.
```

```python
from statsmodels.stats.power import TTestIndPower

effect_size = 0.2  # small effect
power = 0.8        # standard
alpha = 0.05

analysis = TTestIndPower()
required_n = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha)
print(f"Recommended sample size per group: {int(required_n)}")
```
